```python

from transformers import AutoTokenizer

model_name = "Qwen/Qwen2.5-0.5B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name)

requests = pd.read_csv("/home/ssn899/Desktop/P2P_LLM/Distributed-Inference-for-Edge-LLMs/datasets/requests.csv")
```

```python
max_len = 0
min_len = 1000000000
for idx,ele in requests.iterrows():
    curr_len = len(tokenizer.encode(ele['text'], add_special_tokens=False))
    max_len = max(max_len,curr_len)
    min_len = min(min_len, curr_len)


print(max_len, min_len)
```
