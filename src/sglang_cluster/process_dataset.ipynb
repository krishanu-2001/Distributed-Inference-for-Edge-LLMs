```python
from pathlib import Path

import pandas as pd

requests_path = Path("/home/ssn899/Desktop/P2P_LLM/Distributed-Inference-for-Edge-LLMs/datasets/requests.csv")
num_nodes = 4

df = pd.read_csv(requests_path)

if "subject" not in df.columns:
    raise ValueError("requests.csv must contain a 'subject' column")

# First occurrence of each subject is balanced across nodes 0, 1, 2, 3.
# Every non-first occurrence then follows one global round-robin sequence.
seen_subjects = set()
first_subject_node = 0
round_robin_node = 0
target_nodes = []

for subject in df["subject"]:
    if subject not in seen_subjects:
        target_nodes.append(first_subject_node)
        seen_subjects.add(subject)
        first_subject_node = (first_subject_node + 1) % num_nodes
    else:
        target_nodes.append(round_robin_node)
        round_robin_node = (round_robin_node + 1) % num_nodes

df["target_node"] = target_nodes
df.to_csv(requests_path, index=False)

print(f"Updated {requests_path}")
print(df["target_node"].value_counts().sort_index())
display(df.head())
```
